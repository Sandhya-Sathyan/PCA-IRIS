```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report
df=pd.read_csv('/content/drive/MyDrive/DatasetML/IRIS.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['species']=le.fit_transform(df['species'])
df
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
mm=MinMaxScaler()
X=mm.fit_transform(X)
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
pca=PCA(n_components=2,random_state=1)
X_train=pca.fit_transform(X_train)
X_test=pca.transform(X_test)
X_train
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
y_pred
```

```python
print(classification_report(y_test,y_pred))
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report
```

```python
df1=pd.read_csv('/content/drive/MyDrive/DatasetML/breast-cancer.csv')
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['diagnosis']=le.fit_transform(df1['diagnosis'])
df1
```

```python
X=df1.iloc[:,2:]
X
```

```python
df1.drop('id',axis=1,inplace=True)
```

```python
y=df1.iloc[:,-31]
y
```

```python
mm=MinMaxScaler()
X=mm.fit_transform(X)
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.3,random_state=1)
```

```python
pa=PCA(n_components=30,random_state=1)
X_train=pa.fit_transform(X_train)
X_test=pa.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
y_pred
```

```python
print(classification_report(y_test,y_pred))
```
